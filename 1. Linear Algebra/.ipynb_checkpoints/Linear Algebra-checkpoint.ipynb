# Ex1: Vectors

```python
import numpy as np
from numpy.linalg import norm
import math
```

```python
u = np.array([1,3,4,6,3,4,5])
v = np.array([4,7,1,2,3,8,9])
```

__1. Funtion cal_vectors(op, u, v)__
<br> op: +, -, *, /, "dot"
<br> u, v: vectors

```python
def cal_vectors(op, u, v):
    result = None
    if op == "+":
        result = u + v
    elif op == "-":
        result = u - v
    elif op == "*":
        result = u * v
    elif op == "/":
        result = u / v
    elif op == "dot":
        result = u.dot(v)
    else:
        result = None
    return result
```

```python
result = cal_vectors('dot',u,v)
result
```

```python
result = cal_vectors('+',u,v)
result
```

__2. Vector Norm__
<br> Function cal_vector_norm(t,v) with t = ['L1','L2','Max'] and v is vector

```python
def cal_vector_norm(t, v):
    result = None
    if t == "L1":
        result = norm(v,1)
    elif t == "L2":
        result = norm(v)
    elif t == "Max":
        result = norm(v, math.inf)
    else:
        result = None
    return result
```

```python
cal_vector_norm('L2',v)
```

# Ex 2: Matrices

__1. Func cal_matrices(op,m1,m2)__
<br> op = +,-,*,/
<br> m1,m2: matrices

```python
def cal_matrices(op, m1, m2):
    result=None
    if op == "+":
        result = m1 + m2
    elif op == "-":
        result = m1 - m2
    elif op == "*":
        result = m1 * m2
    elif op == "/":
        result = m1 / m2
    else:
        result=None
    return result
```

```python
m1 = np.random.randint(1,10,(4,5))
m1
```

```python
m2 = np.random.randint(5,20,(4,5))
m2
```

```python
result = cal_matrices("+", m1, m2)
result
```

```python
result = cal_matrices("*", m1, m2)
result
```

__2. Matrix-Vector Multiplication__

```python
def create_matrix(m,n):
    lst=[]
    for i in range(m):
        lst_sub=[]
        for j in range(n):
            s='M[' + str(i+1) + ',' + str(j+1) +']:'
            x= eval(input(s))
            lst_sub.append(x)
        lst.append(lst_sub)
    return np.array(lst)
```

```python
def create_vector(m):
    lst=[]
    for i in range(m):
        s='v[' + str(i+1) +']:'
        x= eval(input(s))
        lst.append(x)
    return np.array(lst)
```

```python
A = create_matrix(2,3)
```

```python
v = create_vector(3)
```

```python
A.dot(v)
```

__3. Matrix-Vector Multiplication__
<br> with random indexes

```python
import random
```

```python
def create_matrix_random(m,n,start,end):
    lst=[]
    for i in range(m):
        lst_sub=[]
        for j in range(n):
            s= random.randint(start, end+1)
            lst_sub.append(s)
        lst.append(lst_sub)
    return np.array(lst)
```

```python
def create_vector_random(m,start,end):
    lst=[]
    for i in range(m):
        s=random.randint(start, end+1)
        lst.append(s)
    return np.array(lst)
```

```python
A = create_matrix_random(2,3,1,10)
A
```

```python
v = create_vector_random(3,5,15)
v
```

```python
A.dot(v)
```

# Ex5: Sparse Matrix

__1. Sparsity__
<br> Create A(5,5) with random from -10 to 5
<br> Replace A(<0) = 0
<br> Sparse matrix S from A
<br> Calculate Sparsity

```python
A = np.random.randint(-10,5,(5,5))
A
```

```python
A[A<0] = 0
A
```

```python
from scipy.sparse import csr_matrix
S = csr_matrix(A)
print(S)
```

```python
sparsity = 1 - np.count_nonzero(A)/A.size
sparsity
```

__3. Density value__
<br> Density value = 1- Sparsity
<br> Create sparse matrix S(5,5) with density = 0.25

<br>

```python
import scipy.sparse as sparse
S = sparse.random(5,5, density = 0.25)
print(S)
```

```python
A = S.toarray()
A
```

```python
plt.spy(A, markersize=10)
```
